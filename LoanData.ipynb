```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#This dataset contains loan characteristics of a bank's customers 
#The goal is to clean the dataset, create a model to predict whether or not a loan default occurs, and
#find the characteristics that make for a good loan recipient
```

```python
#Creating the dataset
data = pd.read_csv('LoansTrainingSet.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
#Get rid of non-integer units in series
for ch in ['$', ',']:
    data['Monthly Debt'] = [i.replace(ch, '') for i in data['Monthly Debt']]
```

```python
data['Monthly Debt'] = [float(i) for i in data['Monthly Debt']]
```

```python
data['Monthly Debt']
```

```python
data['Loan Status'].unique()
```

```python
#Change 'Loan Status' unique values to integers
data1['Loan Status'] = data['Loan Status'].replace('Fully Paid', 1).replace('Charged Off', 0)
```

```python
data1['Loan Status']
```

```python
#The columns 'Loan Status', 'Home Ownership', and 'Purpose' are all categorical 
#Here we create dummy variables for them so Python can interpret them
#Dummy variables represent categorical data as numeric
data2 = pd.merge(data1, (pd.get_dummies(data['Term'], drop_first = True)),left_index = True, right_index = True) 
```

```python
data3 = pd.merge(data2, (pd.get_dummies(data['Home Ownership'], drop_first = True)), left_index = True, right_index = True)
```

```python
data4 = pd.merge(data3, (pd.get_dummies(data['Purpose'], drop_first = True)), left_index = True, right_index = True)
```

```python
del data4['Term']
del data4['Home Ownership']
del data4['Purpose']
```

```python
data4.head()
```

```python
data4.info()
```

```python
data4.corr()
```

```python
#Next we deal with missing values for 'Bankruptcies' and 'Tax Liens'
#Correlation map shows both variables are correlated with 'Number of Credit Problems'
#We fill missing values depending on how many credit problems that individual has had
```

```python
#Knowing how many credit problems someone has can help us predict how many bankruptcies they have had
data4.groupby('Number of Credit Problems')['Bankruptcies'].mean()
```

```python
for i in data4[data4['Bankruptcies'].isnull()].index:
    data4['Bankruptcies'][i] = data['Bankruptcies'][data4['Number of Credit Problems'] == data4.ix[i]['Number of Credit Problems']].mean()
```

```python
for i in data4[data4['Tax Liens'].isnull()].index:
    data4['Tax Liens'][i] = data4['Tax Liens'][data4['Number of Credit Problems'] == data4.ix[i]['Number of Credit Problems']].mean()
```

```python
#Each Loan is suppose to be unique, therefore duplicates are dropped
data5 = data4.drop_duplicates(['Loan ID'], keep = 'last')
```

```python
#Max credit score an individual can have is 800 based off of this particular credit scoring system. 
#It appears an extra zero was added to these credit scores
data5['Credit Score'][data5['Credit Score'] > 800]
```

```python
#We divide these values by ten to get rid of the extra zero
#This is done through list comprehension
data5['Credit Score'][data5['Credit Score'] > 800] = [i/10 for i in data5['Credit Score'][data5['Credit Score'] > 800]]
```

```python
#There are now no more values above 800 
data5['Credit Score'].describe()
```

```python
#Some individuals had absurdly high loan amounts yet low income
#Delete outliers in 'Current Loan Amount'
data6 = data5.drop(data5[data5['Current Loan Amount'] > 1000000].index)
```

```python
#Next 'Years in current job' is cleaned
data['Years in current job']
```

```python
h = []
[h.append(i.split(' ')[0]) for i in data6['Years in current job'] if type(i) == str]
len(h)
```

```python
h = [i.replace('<', '.5') for i in h]
h = [i.replace('n/a', '0') for i in h]
h = [i.replace('10+', '10') for i in h]
h = [float(i) for i in h]
h
data6['Years in current job'] = h
```

```python
hh = list(data6['Maximum Open Credit'])
hh
```

```python
#Now 'Maximum Open Credit is cleaned by getting rid of strings
hh = [str(i).replace('#VALUE!', '0') for i in hh]
hh = [float(i) for i in hh]
data6['Maximum Open Credit'] = hh
```

```python
data6['Maximum Open Credit'] = [float(i) for i in data6['Maximum Open Credit']]
```

```python
data6.info()
```
